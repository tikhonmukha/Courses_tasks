```python
df = pd.read_csv('COVID-19 Survey Student Responses.csv')
```

```python
df.head()
```

```python
df = df.assign(sleep_normal=['yes' if i>=7 and i<=8 else 'no' for i in df['Time spent on sleep']])
```

```python
df.head()
```

```python
print(df.loc[df['sleep_normal'] == 'no', 'sleep_normal'].count())
```

```python
df['Time spent on TV'] = df['Time spent on TV'].replace(['n','N','No tv',' '], 0)
```

```python
print(df['Time spent on TV'].dtypes)
```

```python
df = df.assign(sleep_normal_cat=['normal' if i>7 else 'not normal' for i in df['Time spent on sleep']])
```

```python
df = df.assign(media_normal_cat=['normal' if i<2 else 'not normal' for i in df['Time spent on social media']])
```

```python
cross = pd.crosstab(df.sleep_normal_cat, df.media_normal_cat)
cross.iloc[1,1]
```

```python
print(stats.chisquare([cross.iloc[0,0], cross.iloc[0,1], cross.iloc[1,0], cross.iloc[1,1]]).pvalue)
```

```python
top_media = df['Prefered social media platform'].describe()['top']
```

```python
mean_time = df.loc[df['Prefered social media platform'] == top_media, 'Time spent on social media'].mean()
```

```python
print(round(mean_time, 2))
```

```python
media_time = round(df[['Prefered social media platform', 'Time spent on social media']].groupby(by='Prefered social media platform').mean().sort_values('Time spent on social media', ascending=False),2)
```
