```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('PokemonData.csv')
```

```python
df.head()
```

```python
plt.hist(df['Attack'])
plt.savefig('result.png')
```

```python
plt.hist(df['Attack'],alpha=0.5)
plt.savefig('result.png')
```

```python
plt.hist(df['Attack'])
plt.xlabel("Мощность атаки")
plt.ylabel("Количество покемонов")
plt.savefig('result.png')
```

```python
plt.scatter(df['Attack'], df['Defense'], alpha=0.5)
plt.show()
```

```python
df.groupby('Legendary')['Type1'].value_counts().unstack(0).plot(kind = 'bar')
```

```python
df['Type1'].value_counts().plot(kind = 'bar', title="Легендарные покемоны по типам в сравнении с обычными", xlabel="Тип покемонов", ylabel="Количество")
plt.show()
```

```python
df_usual = df[df.Type2.isnull()]
df_unusual = df[df.Type2.notnull()]
plt.hist(df_usual['Attack'], label = 'Обычная атака')
plt.hist (df_unusual['Attack'], label = 'Специальная атака')
plt.xlabel("Мощность атаки")
plt.ylabel("Количество покемонов")
plt.legend()
plt.show()
```
